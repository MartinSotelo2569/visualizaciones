```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_forbes = pd.read_excel('/content/drive/MyDrive/Bootcamp Data Analytics/Ejercicios de la plataforma Coding Dojo/Data Set_Forbes/forbes.xlsx')
df_forbes
```

```python
df_forbes.info()
```

```python
print(df_forbes.columns)
```

```python
df_forbes.columns = df_forbes.columns.str.strip()
print(df_forbes.columns)
```

```python
fig = px.bar(df_forbes, x='NAME', y='TOTAL EARNINGS', title='Ganancias Totales por Deportista')
fig.show()
```

```python
fig = px.pie(df_forbes, names='SPORT', title='Distribuci√≥n de Deportes entre Atletas')
fig.show()
```

```python
fig = px.scatter(df_forbes, x='ON-THE-FIELD EARNINGS', y='OFF-THE-FIELD-EARNINGS',
                 color='SPORT', title='Ganancias en el Campo vs. Ganancias Fuera del Campo')
fig.show()
```

```python
fig = px.bar(df_forbes, x='SPORT', y=['ON-THE-FIELD EARNINGS', 'OFF-THE-FIELD-EARNINGS'],
             barmode='group', title='Comparaci√≥n de Ganancias en el Campo y Fuera del Campo por Deporte')
fig.show()
```

```python
df_soccer = df_forbes[df_forbes['SPORT'] == 'Soccer']
fig = px.bar(df_soccer, x='NAME', y='TOTAL EARNINGS', title='Ganancias Totales por Deportista - F√∫tbol')
fig.show()
```

```python
fig = px.scatter(df_soccer, x='ON-THE-FIELD EARNINGS', y='OFF-THE-FIELD-EARNINGS',
                 color='SPORT', title='Ganancias en el Campo vs. Ganancias Fuera del Campo - F√∫tbol')
fig.show()
```

```python
fig = px.bar(df_soccer, x='NAME', y=['ON-THE-FIELD EARNINGS', 'OFF-THE-FIELD-EARNINGS'],
             title='Comparaci√≥n de Ganancias en el Campo y Fuera del Campo por Deportista - F√∫tbol',
             barmode='stack')
fig.show()
```

```python
df_basketball = df_forbes[df_forbes['SPORT'] == 'Basketball']
fig = px.bar(df_basketball, x='NAME', y='TOTAL EARNINGS', title='Ganancias Totales por Deportista - Baloncesto')
fig.show()
```

```python
fig = px.scatter(df_basketball, x='ON-THE-FIELD EARNINGS', y='OFF-THE-FIELD-EARNINGS',
                 color='SPORT', title='Ganancias en el Campo vs. Ganancias Fuera del Campo - Baloncesto')
fig.show()
```

```python
fig = px.bar(df_basketball, x='NAME', y=['ON-THE-FIELD EARNINGS', 'OFF-THE-FIELD-EARNINGS'],
             title='Comparaci√≥n de Ganancias en el Campo y Fuera del Campo por Deportista - Baloncesto',
             barmode='stack')
fig.show()
```
